```python
#Although this toolset was developed to simplify data management, it proved preferable to simply carry out processing 
#semi-manually in excel

import pandas as pd
import jellyfish as jf

def comPairison(ini, sec, th):
    #code updated from https://miguendes.me/python-compare-strings
    #For a given threshold of similarity in keystrokes, determines if two strings belong matched together.
    #provide ini and sec in the form of strings, th in the form of an integer.
    
    return jf.damerau_levenshtein_distance(ini,sec) <= th

def sexIngest(file, product):
    #takes a census data CSV with gender delineated data and outputs a version with only total values. Parameters as strings.
    f = pd.read_csv(file)
    for i in list(f):
        for u in list(f):
            if i != u:
            #for the purposes of gendered census data, the difference will be 3 between Male and Female.
                if comPairison(i,u,3) == True:
                    q = i + ('_combinedSexes')
                    f[q] = f[i] + f[u] 
    f.to_csv(product)
    return f
```
